## Model BUILDING

```python
from sklearn.model_selection import train_test_split
train , test = train_test_split(iris,test_size = 0.2)
```

```python
train_x = train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
train_y = train[['Species']]
```

```python
test_x = test[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
test_y = test[['Species']]
```

## DESCISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(train_x,train_y)
```

```python
y_predict = model.predict(test_x)
```

```python
## Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm = confusion_matrix(y_predict,test_y)
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel('y_predict')
plt.ylabel('test_y')
plt.show()
```

```python
print(classification_report(y_predict,test_y))
```

```python
print("Accuracy Score:",accuracy_score(y_predict,test_y))
print('Model Score:',model.score(test_x,test_y))
```

```python
x = np.array([[5.9,3.0,5.1,1.8 ],[5.0,3.6,1.4,0.2]])
prediction = model.predict(x)
print("Prediction of Species: {}".format(prediction))
```
