```python
highest_unemployment = dataset.loc[dataset[' Estimated Unemployment Rate (%)'].idxmax()]
print("Highest Unemplyment Rate: \n")
print(highest_unemployment)
```

```python
Lowest_unemployment = dataset.loc[dataset[' Estimated Unemployment Rate (%)'].idxmin()]
print("Lowest Unemplyment Rate: \n")
print(Lowest_unemployment)
```

Data Visualisation

```python
freq = dataset['Region'].value_counts()
freq2 = dataset['Area'].value_counts()
freq2.plot(kind = 'pie',startangle = 0)
plt.title("Unemployment in Rural and Urban Areas")
plt.legend(freq2)
plt.show()
freq.plot(kind = 'pie',startangle = 0)
plt.title("Unemployment in States")
plt.legend(freq)
plt.show()
```

```python
sns.pairplot(data=dataset, hue="Region", vars=[' Estimated Unemployment Rate (%)',
       ' Estimated Employed', ' Estimated Labour Participation Rate (%)'],height=3, aspect=1)
```

```python
sns.pairplot(data=dataset, hue="Area", vars=[' Estimated Unemployment Rate (%)',
       ' Estimated Employed', ' Estimated Labour Participation Rate (%)'],height=3, aspect=1)
```

```python
plt.figure(figsize =(15,12))
sns.countplot(x = 'Region',data = dataset,orient = 'h')
plt.xticks(rotation = 65)
plt.ylabel('Estimated Unemployment Rate (%)')
```

```python
avg = dataset.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
avg.plot(kind='bar')
plt.title("Average Unemplyment Rate by Region")
plt.xlabel('Region')
plt.ylabel('Average Unemplyment Rate (%)')
plt.show()
```
