```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
car = ['fueltype', 'aspiration','doornumber', 'carbody', 'drivewheel', 'enginelocation','enginetype',
       'cylindernumber','fuelsystem']
for i in car:
    label_encoder = LabelEncoder()
    df[i+'_label_encoder'] = label_encoder.fit_transform(df[i])
df.drop(car,axis = 1,inplace= True)
df
```

```python
x = df.drop(columns = ['price','CarName'],axis=1)
y = df['price']
print(x.shape,y.shape)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
print("Training Data")
print('X_train')
X_train
```

```python
print("Y_train")
Y_train
```

```python
print('Testing Data')
print('X_test')
X_test
```

```python
print("Y_test")
Y_test
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
model1.fit(X_train,Y_train)
```

```python
training_data_prediction = model1.predict(X_train)
testing_data_prediction = model1.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R squared error :",error_score)
```

```python
error_score = metrics.r2_score(Y_test,testing_data_prediction)
print("R squared error :",error_score)
```

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price v/s Predicted Price')
```

```python
plt.scatter(Y_test,testing_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price v/s Predicted Price')
```
